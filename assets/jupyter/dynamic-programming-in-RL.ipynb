```python
import numpy as np

gamma = 0.9
R = np.array([1., 1., 0., 0., 1., 4., 0., 0., 0., 2.], dtype=np.float32)
R
```

```python
P = np.random.randn(10, 10)
P = np.abs(P)
P = P / P.sum(axis=1, keepdims=True)
P
```

```python
I = np.eye(N=10)
V = np.linalg.inv(I - gamma * P) @ R
np.set_printoptions(suppress=True)
V.round(2)
```
