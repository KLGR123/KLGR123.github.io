```python
import json
import numpy as np

x = '{"train":[[2,0,0,0],[3,1,0,0],[0,0,2,1],[0,1,3,1]],"test":[[1,0,0],[0,1,2]]}'
x = x.strip()
data = json.loads(x)

train = np.array(data['train'], dtype=np.int32)
test = np.array(data['test'], dtype=np.int32)

X_train = train[:, :-1]
y_train = train[:, -1]

n0 = np.sum(y_train == 0)
N = len(y_train)
pi0 = n0 / N
pi1 = 1 - pi0

X_train[y_train == 0], X_train[y_train == 1]
```

```python
X_train[y_train == 0].sum(axis=0)
```

```python
count_0 = X_train[y_train == 0].sum(axis=0)
count_1 = X_train[y_train == 1].sum(axis=0)

logprob_0 = np.log((count_0 + 1) / np.sum(count_0 + 1, axis=0))
logprob_1 = np.log((count_1 + 1) / np.sum(count_1 + 1, axis=0))

logprob_0, logprob_1
```

```python
score_0 = test @ logprob_0 + np.log(pi0)
score_1 = test @ logprob_1 + np.log(pi1)
pred = (score_1 >= score_0).astype(int)
pred.tolist()
```
