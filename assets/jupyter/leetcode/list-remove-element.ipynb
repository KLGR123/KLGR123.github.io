27. 移除元素

https://leetcode.cn/problems/remove-element/description/?envType=study-plan-v2&envId=top-interview-150

这道题还是双指针，没话说，第一次自己做对了。整体思路就是，一个读头从前往后，一个写头从后往前；一旦遇到要移除的数字，将该元素处换成写头所在位置的元素，写头随便写个 token 占位就行了，比如这里 -1 没问题。

但是提交了三四次，边界条件总是抓不准。共有三个小坑
- 首先，对于 nums 长度为 0 和 1 的情况要先单独写边界，这个不难写但是老忘记；
- return 的是 k + 1 而不是 k，因为 k 是下标，返回的是长度，当出现下标转长度时就 + 1，反之 - 1；
- 最难的一个边界，最不好想的就是如果写头的位置也是要移除的数字 val 时的情况，这个时候需要先 while 到不是 val 的位置，处理起来很容易，但是想的时候容易忽略。

```python
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return 0 if nums[0] == val else 1

        i, k = 0, len(nums) - 1
        while True:
            if i == len(nums) - 1:
                return k + 1

            while nums[k] == val:
                nums[k] = -1
                k -= 1

            if nums[i] == val:
                nums[i] = nums[k]
                nums[k] = -1
                k -= 1  
                
            i += 1
```

26. 删除有序数组中的重复项

输入是 nums = [0,0,1,1,1,2,2,3,3,4]；输出是 (5, nums = [0,1,2,3,4,_,_,_,_,_])。

这题也是一样，双指针，读头往后顺序遍历，写头位置的值一旦和读头不一样了，往后走一个并写入新的值。这题一遍过。

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i, j = 0, 1
        for j in range(len(nums)):
            if nums[i] != nums[j]:
                i += 1
                nums[i] = nums[j]
        return i + 1
```

80. 删除有序数组中的重复项 II

https://leetcode.cn/problems/remove-duplicates-from-sorted-array-ii/description/?envType=study-plan-v2&envId=top-interview-150

我的思路还和删除有序数组 I 的想法一样，就是多维护一个变量，这个变量负责记录重复了几次，但是很明显这个变量没有尽到它应尽的义务，在遇到 [1,1,1,1] 或更长的时候时就不对了。这是一个粗心的错误，很明显 r 的归零不应该在 r == 1 时，而是在下一次出现不同时。

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i, r = 0, 0
        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                i += 1
                nums[i] = nums[j]
            else:
                if r < 1:
                    r += 1
                    i += 1
                    nums[i] = nums[j]
                else:
                    r = 0
        return i + 1
```

改进后会发现，问题解决了。

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i, r = 0, 0
        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                i += 1
                nums[i] = nums[j]
                r = 0
            else:
                if r < 1:
                    r += 1
                    i += 1
                    nums[i] = nums[j]
        return i + 1
```
