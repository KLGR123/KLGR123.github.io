69. x 的平方根 

https://leetcode.cn/problems/sqrtx/description/?envType=study-plan-v2&envId=top-interview-150

用二分查找实现平方根的计算。

这里比较难找的，也是容易犯错的就是 left right 的终止边界。如果是在 left + 1 == right 时就停止，那么可能会出现当倒数第二次更新时，left 如果小了则还需要更新 left = mid + 1，此时 left == right 但还没有计算此时 left 和 right 共同指向的位置的平方。也就是说，少了一次计算机会。

另一个小坑就是 ans 的更新时机，应该是在下界更新的时候，因为是下取整。最后就是，ans 更新的时候应该更新为 mid 而不是 left 即 mid + 1。

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        left = 0
        right = x
        ans = 0
        if x < 2:
            return x
        while left <= right:
            mid = (left + right) // 2
            if mid ** 2 == x:
                return mid
            elif mid ** 2 < x:
                left = mid + 1
                ans = mid # not left
            else:
                right = mid - 1
        return ans
```

如果规范化地写，那么就是左闭右开吗？这道题还真不是，注意看如下写法，对于 8 返回的是 3 而不是 2；本题的要求是结果只保留整数部分，小数部分将被舍去，因此是下取整而不是上取整数。可以说，左闭右开写法在处理连续区间问题时，天然地会上取整。

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x
        l, r = 0, x
        while l < r:
            m = (l + r) // 2
            if m ** 2 < x:
                l = m + 1
            else:
                r = m
        return l
    
solution = Solution()
print(solution.mySqrt(10))
```

如果想要变成下取整，首先不可能粗暴地返回 l - 1，那对于平方数是无效的；因此，再考虑平方数的情况，我们要对左闭右开算法做的第二处连贯调整，即把 `m ** 2 == x` 的情况留给左边界；如此一来，平方数返回的即为 m + 1 - 1 = m。

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x
        l, r = 0, x
        while l < r:
            m = (l + r) // 2
            if m ** 2 <= x:
                l = m + 1
            else:
                r = m
        return l - 1
    
solution = Solution()
print(solution.mySqrt(8)) 
```

左闭右开找下界：
- 区间始终表示可能答案；
- `< target`，丢左边：l = m + 1；
- `>= target`，收右边：r = m；
- l == r 时，l 就是答案。

```python
from typing import List

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums)
        while l < r:
            m = (l + r) // 2
            if nums[m] < target:
                l = m + 1
            else:
                r = m
        return l
```

以下还有一个调包的方式，仅供参考，bisect_left 的使用。

```python
from bisect import bisect_left
from typing import List

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        return bisect_left(nums, target)
```

162. 寻找峰值

https://leetcode.cn/problems/find-peak-element/description/?envType=study-plan-v2&envId=top-interview-150

峰值元素是指其值严格大于左右相邻值的元素。

开始我的数学思路作为一种直觉，告诉我应该先做差分。然而二分法比这想法美妙的多。

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        if len(nums) <= 1:
            return 0
        l, r = 0, len(nums) - 1
        while l < r:
            mid = (l + r) // 2
            if nums[mid + 1] > nums[mid]:
                l = mid + 1
            else:
                r = mid
        return l
    
solution = Solution()
nums = [9,7,3,7,8]
solution.findPeakElement(nums)
```
