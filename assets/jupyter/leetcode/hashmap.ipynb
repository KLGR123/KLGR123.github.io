383. 赎金信

https://leetcode.cn/problems/ransom-note/description/?envType=study-plan-v2&envId=top-interview-150

给你两个字符串：ransomNote 和 magazine ，判断 ransomNote 能不能由 magazine 里面的字符构成。

以下是我的最开始的写法，我觉得用字典就是哈希，但这样做似乎在时间上比较慢。

```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        hashmap = {}
        for s in magazine:
            if s in hashmap:
                hashmap[s] += 1
            else:
                hashmap[s] = 1
        for s in ransomNote:
            if s in hashmap:
                if hashmap[s] == 0:
                    return False
                hashmap[s] -= 1
            else:
                return False
        return True
```

如下是比较优雅的写法，要学会用 cnt.get(ch, 0)。

```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        cnt = {}
        for ch in magazine:
            cnt[ch] = cnt.get(ch, 0) + 1

        for ch in ransomNote:
            if cnt.get(ch, 0) == 0:
                return False
            cnt[ch] -= 1
        return True
```

205. 同构字符串

给定两个字符串 s 和 t ，判断它们是否是同构的。

如果 s 中的字符可以按某种映射关系替换得到 t ，那么这两个字符串是同构的。

首先是我如下这种写法，是有问题的，显而易见没有考虑双射。

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        cnt = {}
        for s_ch, t_ch in zip(s, t):
            if s_ch in cnt:
                t_done = cnt[s_ch]
                if t_done != t_ch:
                    return False
            else:
                print(cnt)
                cnt[s_ch] = t_ch
        return True

solution = Solution()
solution.isIsomorphic("badc", "baba")
```

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        last_s = {}
        last_t = {}
        for i, (a, b) in enumerate(zip(s, t)):
            if last_s.get(a, -1) != last_t.get(b, -1):
                return False
            last_s[a] = i
            last_t[b] = i
        return True
```
