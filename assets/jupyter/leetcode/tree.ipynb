单调栈，但对于不同路径。

```python
n = 7
x = [0] + [7, 1, 2, 3, 4, 5, 6]
g = [[] for _ in range(n + 1)]

pairs = [
    (1, 2),
    (2, 3),
    (3, 4),
    (5, 4),
    (5, 6),
    (6, 7)
]
for u, v in pairs:
    g[u].append(v)
    g[v].append(u)

g
```

```python
extra = [[] for _ in range(n + 1)]
stack = []

def dfs(u, parent):
    j = -1
    for val, node in reversed(stack):
        if val > x[u]:
            j = node
            break
    if j != -1:
        extra[u].append(j)
        extra[j].append(u)

    removed = []
    while stack and stack[-1][0] <= x[u]:
        removed.append(stack.pop())
    
    stack.append((x[u], u))
    for v in g[u]:
        if v != parent:
            dfs(v, u)
    stack.pop()

    while removed:
        stack.append(removed.pop())

dfs(1, 0)
```

```python
adj = [[] for _ in range(n + 1)]
for u in range(1, n + 1):
    for v in g[u]:
        adj[u].append(v)
    for v in extra[u]:
        adj[u].append(v)
adj
```

求最短路径 d，可以用 BFS 解决。图变成原本树的边，加上新的边。

```python
from collections import deque

distance = [-1] * (n + 1)
q = deque([1])
distance[1] = 0

while q:
    u = q.popleft()
    for v in adj[u]:
        if distance[v] == -1:
            distance[v] = distance[u] + 1
            q.append(v)

distance[1:]
```
