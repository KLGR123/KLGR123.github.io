274. H 指数

https://leetcode.cn/problems/h-index/description/?envType=study-plan-v2&envId=top-interview-150

做法 1 是排序，最常用，代码最短，把引用次数从大到小排序。如果排完后第 i 篇论文的引用 >= i，说明至少有 i 篇论文各自引用都 ≥ i，可以更新 h。可以看到，任何桶排序问题，实际上都可以通过直接 sort 解决，桶只不过是将索引**压缩**了。

这里的坑其实是关于反直觉。先排序后，然后从大到小地取出 citation；这时，我们在乎的是一旦某个位置 i 的 citation 数量开始小于 i + 1（注意是小于，不是小于等于，等于时不能保证下一个 citation 是否还等于，如果还等于不应该退出而是继续增大 h-index），注意等号应该放在不退出的条件，而不是退出条件中。第二个坑是，你要更新的 h-index 不是 c 而是 i + 1，是篇数而不是当前篇的引用数。

```python
from typing import List

class Solution:
    def hIndex(self, citations: List[int]) -> int:
        citations.sort()
        h = 0
        for i, c in enumerate[int](citations[::-1]):
            if i + 1 <= c:
                h = i + 1
            else:
                break
        return h
```

用桶的关键观察是，h 最大不会超过论文总数 n。把每篇论文的引用次数 c 归到桶里，若 c >= n，都算进 bucket[n]（因为超过 n 对 h 没区别），否则进 bucket[c]；然后从大到小累加论文数 papers，第一次满足 papers >= h 的 h 就是答案。

这里的坑和如上同样是对偶的，第一个是 [0] * (n + 1) 也就是 bucket 的长度实际上是 n + 1 的；第二个是 return 的是 i 而不是 papers，也就是要的是 index。

```python
from typing import List

class Solution:
    def hIndex(self, citations: List[int]) -> int:
        n = len(citations)
        bucket = [0] * (n + 1)

        for c in citations:
            if c >= n:
                bucket[n] += 1
            else:
                bucket[c] += 1

        papers = 0
        for i in range(n, -1, -1):
            papers += bucket[i]
            if papers >= i:
                return i # not papers but i
```
