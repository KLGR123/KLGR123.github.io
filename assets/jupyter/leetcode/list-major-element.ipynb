169. 多数元素

https://leetcode.cn/problems/majority-element/description/?envType=study-plan-v2&envId=top-interview-150

最好想的自然是哈希，遍历一遍数组，然后统计频率。这里还有一个小点，即遍历到某个元素并计数时的 cnt.get(num, 0) 可以自动实现 if 不存在的情况，以及当时如果就大于 half 就返回。

```python
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        cnt = {}
        half = len(nums) // 2

        for num in nums:
            cnt[num] = cnt.get(num, 0) + 1
            if cnt[num] > half:
                return num
```

第二好想的是排序后取中位数即可，复杂度是 o(nlogn)。其实是很快的，但还有时间复杂度是 o(n) 的算法。

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        nums.sort()
        return nums[len(nums) // 2]
```

如下即一种最优的多数元素算法，它的名字也是它的发明人名字，叫做 **Boyer–Moore 投票算法**。在看过如下 AI 给出的答案后，我明白了这是怎么一回事。可以想像这样一个场景，有若干士兵，每个士兵都属于一个队伍（这里元素值相同的元素即属于一个队伍），这时国王希望知道哪个队伍的人最多。对于计算机来说，比较仁慈的、但需要额外记录空间的做法是，统计哈希（相当于大臣拿了一个小本本额外记录）；但如果国王强制要求不需要额外空间，连脑子也不要用（国王懒得动脑），那该如何？比较暴力却不需要额外空间的做法就是，让士兵排好队，互相厮杀，如果是同队伍则站在一起，否则同归于尽、战死角斗场。最后剩下的，一定是多数者的一队的某一员。

这在数学上也很好想，即假若存在一个元素其数量大于 n // 2，那么用剩下所有的元素去与之“消消乐”，也一定会剩下的。但是前提是要保证确实存在这样一种元素，这是题目的假设。

```python
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        candidate = None
        count = 0

        for x in nums:
            if count == 0:
                candidate = x
                count = 1
            elif x == candidate:
                count += 1
            else:
                count -= 1

        return candidate
```

```python
solution = Solution()
solution.majorityElement(nums=[1, 2, 3, 1]), solution.majorityElement(nums=[1, 2, 1, 3, 1])
```

我的写法最开始犯小错是在 candidate 的初始化，即想成了 nums[0]，并使循环从 nums[1:] 开始。这样做的问题是，如果 nums[0] 就是那个最多元素的时候，count 会有 1 个单位的相位差，因此会错。

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        winner, pmf = None, 0 
        for num in nums:
            if num == winner:
                pmf += 1
            elif pmf == 0:
                winner = num
                pmf += 1
            else:
                pmf -= 1

        return winner
```
