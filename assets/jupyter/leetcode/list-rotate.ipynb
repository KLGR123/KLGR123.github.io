189. 轮转数组

https://leetcode.cn/problems/rotate-array/description/?envType=study-plan-v2&envId=top-interview-150

这道题要求用三种做法试一试，实际上不用说咱们也是要高要求的。最简单好想的自然是直接切割，然后重新拼接数组，如下。

错了的坑有
- k 需要先 %= 数组的总长，确保不会超出 index 范围，这种问题以后一旦遇到，比如题目或者自己的算法中间产生了某个变量，用于操纵数组索引，则必须考虑范围限制；
- 最后这里，nums 是原地操作，且注意是先 nums[-k:]。

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k %= n
        nums[:] = nums[-k:] + nums[:-k]
```

或可写成额外数组空间，本质上是一样的。

这里错了的坑是，忘记算 offset 了；又是上面说的问题，当 i + k 超出 n - 1 了，就说明回到数组开头了。

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k %= n
        res = [0] * n
        for i, num in enumerate(nums):
            offset = (i + k) % n
            res[offset] = num
        nums[:] = res
```

最后，让我们理解数组平移 k 个单位的终极实现，以及它为什么是一种旋转。

如下就不画图了，但是可以通过我这里所说的步骤进行作图，就可以明晰为什么**一次 global 旋转 + 两次 local 旋转 = 一次平移**。首先将数组接成环，并将 0 和 k % n 的位置标上不同的记号，然后将如上的平移和旋转操作都进行一遍即可。

如下，实现中的坑有
- 边界条件直接 return，加快速度；
- rev() 操作函数的实现。

如上圆环旋转算法，即彻底明确了如何平移（旋转）数组。

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k %= n

        if k == 0:
            return
        
        def rev(l: int , r: int):
            while l < r:
                nums[l], nums[r] = nums[r], nums[l]
                l += 1
                r -= 1
        
        rev(0, n - 1)
        rev(0, k - 1)
        rev(k, n - 1)
```

如果这里不是向右移动，而是向左移动，又该如何？或者说不是顺时针，而是逆时针旋转呢？实际上，只需要把 k - 1 变成 n - k + 1。
