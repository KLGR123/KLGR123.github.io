快速排序。时间复杂度是平均时间复杂度为 `o(nlogn)`，是实际应用中最高效的排序算法之一。

```python
import time

start = time.time()

arr = [9, 1, 8, 2, 7, 3, 4, 6, 5]

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

quick_sort(arr)
time.time() - start
```

原地快排。

```python
start = time.time()
data = [9, 1, 8, 2, 7, 3, 4, 6, 5]

def quick_sort(l, r):
    if l >= r:
        return 
    i = l - 1
    for j in range(l, r):
        if data[j] <= data[r]:
            i += 1
            data[i], data[j] = data[j], data[i]
    data[i + 1], data[r] = data[r], data[i + 1]
    quick_sort(l, i)
    quick_sort(i + 1, r)

quick_sort(0, len(data) - 1)
print("".join([str(d) for d in data]))
time.time() - start
```
