238. 除了自身以外数组的乘积

https://leetcode.cn/problems/product-of-array-except-self/description/?envType=study-plan-v2&envId=top-interview-150

请不要使用除法，且在 O(n) 时间复杂度内完成此题。标准解法是前缀积 + 后缀积，本质是把每个位置左边的乘积和右边的乘积拆开算，再相乘。这里如果第一次做，思路应该是如此：首先想到整体求乘积，然后再分别广播回去，回去的路上用除法。然后看到题目条件是不允许使用除法，据此分析，如果还要求 o(n) 则势必要在多个无嵌套循环中完成某个中间变量的统计，既然不允许用除法，则一定要用乘法（乘除对偶，加减法不可能完成本题）。因此，考虑前缀和后缀乘积变量。

至于为什么题目不让用除法，理由也是显而易见的，即对于 0 元素需要特殊处理。如下是我写的代码，虽然丑陋了一些，但也可以通过。主要坑位有左右侧 range 范围的控制和 1 的补充。

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        left = [0] * (n - 1)
        right = [0] * (n - 1)
        l = 1
        for i in range(n - 1):
            l *= nums[i] 
            left[i] = l
        left = [1] + left # [1, 1, 2, 6]
        r = 1
        for i in range(n - 1, 0, -1): # 3, 2, 1
            r *= nums[i]
            right[n - 1 - i] = r # 0, 1, 2
        right = right[::-1] + [1] # [4, 12, 24, 24] -> [24, 12, 4, 1]
        answer = [0] * n
        for i in range(n):
            answer[i] = left[i] * right[i]
        return answer

solution = Solution()
solution.productExceptSelf([1, 2, 3, 4])
```

你可以在 O(1) 的额外空间复杂度内完成这个题目吗？这里也是自己做出来的，思路比较好想：如果限制空间复杂度，又需要空间存储（理论上不可能不需要任何空间，因为在遍历某位置 i 时绝无可能同时知道其前缀和后缀），那么就需要更好利用 answer 数组。相当于之前是用两个 cache，现在是直接分阶段写入了 answer。

```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        answer = [1] * n
        l = 1
        for i in range(n - 1):
            l *= nums[i]
            answer[i + 1] = l
        r = 1
        for i in range(n - 1, 0, -1):
            r *= nums[i]
            answer[i - 1] *= r
        return answer
```
