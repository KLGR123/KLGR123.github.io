用二分查找实现平方根的计算。

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        left, right = 0, x
        ans = 0
        while left <= right:
            mid = (left + right) // 2
            sq = mid * mid
            if sq <= x:
                ans = mid
                left = mid + 1
            else:
                right = mid - 1
        return ans
```
