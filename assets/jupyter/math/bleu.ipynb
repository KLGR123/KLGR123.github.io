Extract all the length-$n$ grams from sentences to token lists.

```python
from collections import Counter
import math

tokens = "This is a sample text for testing n-gram frequency.".split()
n = 2  # For bigrams
Counter(tuple(tokens[i:i+n]) for i in range(len(tokens) - n + 1))
```

Modify the $n$-gram precision $p_n$.

```python
def get_ngrams(text, n):
    tokens = text.split()
    return Counter(tuple(tokens[i:i+n]) for i in range(len(tokens) - n + 1))

candidate = "This is a test text text text."
references = [
    "This is text a sample text for testing n-gram frequency.",
    "This is a text test text for n-gram frequency."
]
n = 2

cand_ng = get_ngrams(candidate, n)
print(cand_ng)
ref_max = Counter()
for ref in references:
    for gram, cnt in get_ngrams(ref, n).items():
        ref_max[gram] = max(ref_max[gram], cnt)
print(ref_max)
clipped_sum = sum(min(cnt, ref_max[gram]) for gram, cnt in cand_ng.items())
clipped_sum / sum(cand_ng.values()), clipped_sum, sum(cand_ng.values())
```

Brevity Penalty $BP$.

```python
candidate = "This is a test text text text."
references = [
    "This is text a sample text for testing n-gram frequency.",
    "This is a text test text for n-gram frequency."
]

r = min((len(ref) for ref in references), key=lambda l: abs(l - len(candidate)))
print(r, len(references[0]) - len(candidate), len(references[1]))
1.0 if len(candidate) >= r else math.exp(1 - r / len(candidate))
```
